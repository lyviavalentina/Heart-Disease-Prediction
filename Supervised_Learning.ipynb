# **K-nearest neighbors (KNN)**

```python
import sys
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

Heart_Disease_data = pd.read_csv("/content/heart_disease_data.csv", header=0, sep=";")
x = Heart_Disease_data['thalach']
y = Heart_Disease_data['trestbps']
classes = Heart_Disease_data['target']
colors = ['pink' if cls == 0 else 'yellow' for cls in classes]
plt.clf()
# The 'c' argument should be used to specify the colors
plt.scatter(x, y, c=colors)
plt.title('Heart Disease Data')
plt.xlabel('Maximum heart rate (in bpm)')
plt.ylabel('Blood Pressure (in mm Hg)')
plt.show()
```

```python
data = list(zip(x, y))
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(data, classes)
```

```python
#mendifinisikan titik baru
new_x = 185
new_y = 120
new_point = [(new_x, new_y)]

#membuat prediksi
prediction = knn.predict(new_point)

# Create colors list with the same length as x and y
colors = ['pink' if cls == 0 else 'yellow' for cls in classes]
# Append the color for the new point
colors.append('blue' if prediction[0] == 0 else 'black' ) # Append color based on prediction

#membuat sctter plot
plt.clf()
# Scatter plot with x, y, and colors having the same length
plt.scatter(x.tolist() + [new_x], y.tolist() + [new_y], c=colors)
plt.text(x=new_x, y=new_y - 7, s=f"new point, class: {prediction[0]}", ha='center')

plt.title('Heart Disease Data (KNN)')
plt.xlabel('Maximum heart rate (in bpm)')
plt.ylabel('Blood Pressure (in mm Hg)')
```

# **Decision Tree**

```python
# mengimpor library
import sys
import matplotlib

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Membuat data
Heart_Disease_data = pandas.read_csv("/content/tes heart 21 data.csv", sep=";")

# Memilih fitur
features = ['age', 'trestbps', 'thalach', 'oldpeak']

X = Heart_Disease_data[features]
y = Heart_Disease_data['target']

#Membuat descision tree
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

#memvisualisasikan tree
tree.plot_tree(dtree, feature_names=features)

plt.show()
```

```python
print(dtree.predict([[40, 145, 170, 1]]))

print("[1] means 'YES'")
print("[0] means 'NO'")
```
